```python
import pandas as pd


df = pd.read_csv('data/tourism_with_id.csv')



df = df.drop(['Unnamed: 11', 'Unnamed: 12', 'Lat', 'Long', 'Coordinate', 'Time_Minutes'], axis=1)


"""
I want to normalize df['Rating'] data so that the value is between 0 until 5 because the rating value is between 0 until 5. 

I want to change the label become like this:

4.0 , 4.1 ==> 1,
4.2, 4.3 ==> 2,
4.4, 4.5 ==> 3,
4.6, 4.7 ==> 4,
4.8, 4.9, 5.0 ==> 5
"""

# change the label

df['Rating'] = df['Rating'].replace([4.0, 4.1], 0)
df['Rating'] = df['Rating'].replace([4.2, 4.3], 1)
df['Rating'] = df['Rating'].replace([4.4, 4.5], 2)
df['Rating'] = df['Rating'].replace([4.6, 4.7], 3)
df['Rating'] = df['Rating'].replace([4.8, 4.9, 5.0], 4)

# convert data to int

df["Rating"] = df["Rating"].astype(int)

# select only yogyakarta city

df = df[df['City'] == 'Yogyakarta']

# reset index

df = df.reset_index(drop=True)

df['Place_Id'] = df.index + 1


# save the dataframe to csv file

df.to_csv('data/new_data.csv', index=False)


df.head(10)

```

```python
import tensorflow as tf
import numpy as np








category_test = 3

price_test = 1

input_data = np.array([[category_test, price_test]], dtype=np.float32)

interpreter = tf.lite.Interpreter(model_path="model/filtering_model.tflite")

interpreter.allocate_tensors()

input_details = interpreter.get_input_details()

output_details = interpreter.get_output_details()

interpreter.set_tensor(input_details[0]['index'], input_data)

interpreter.invoke()

output_data = interpreter.get_tensor(output_details[0]['index'])

print(output_data)

print(np.argmax(output_data))

# Menampilkan tempat wisata sesuai dengan rating 
for i in range(len(df)):
    if df['Rating'][i] == np.argmax(output_data):
        print(df['Place_Name'][i])
```
